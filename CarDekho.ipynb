```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib notebook
```

```python
df=pd.read_csv('Car details v3.csv')
```

```python
df
```

```python
df.seats.unique()
```

```python
X=df.drop(['name','torque','selling_price'],axis=1)
y=df['selling_price']
```

```python
X.info()
```

```python
df.isnull().sum()
```

```python
X['mileage']=X['mileage'].str.replace(r'( .*)', '').astype('float64')
```

```python
X['engine']=X['engine'].str.replace(r'( .*)', '').astype('float64')
```

```python
X['max_power']=X['max_power'].str.replace(r'( .*)', '')#.astype('float64')
```

```python
X.max_power.unique()
```

```python
X = X.replace(r'^\s*$', np.nan, regex=True)
```

```python
X.info()
```

```python
X['max_power']=X['max_power'].astype('float64')
```

```python
X.info()
```

```python
X['year']=2021-X['year']
```

```python
X
```

```python
print(X['fuel'].unique())
print(X['seller_type'].unique())
print(X['transmission'].unique())
print(X['owner'].unique())

```

```python
# Dummy Variable Traps
X=pd.get_dummies(X,drop_first=True)
```

```python
X.head()
```

```python
X.info()
```

```python
X.isnull().sum()
```

```python
X['mileage'].fillna(X['mileage'].median(),inplace=True)
X['engine'].fillna(X['engine'].median(),inplace=True)
X['max_power'].fillna(X['max_power'].median(),inplace=True)
X['seats'].fillna(X['seats'].median(),inplace=True)
```

```python
X.isnull().sum()
```

```python
X.seats.unique()
```

```python
X=X.astype('float64')
```

```python
X.info()
```

```python
import seaborn as sns
plt.figure(figsize=(12,10))
sns.heatmap(X.corr(),annot=True)
plt.tight_layout()
```

```python
# feature Imoprtance
from  sklearn.ensemble import ExtraTreesRegressor
model=ExtraTreesRegressor()
model.fit(X,y)
```

```python
model.feature_importances_
```

```python
plt.style.use('seaborn')
plt.barh(X.columns,model.feature_importances_)
plt.tight_layout()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=1)
```

```python
X_train.shape,X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
mod=LinearRegression()
mod.fit(X_train,y_train)
mod.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
r=RandomForestRegressor(random_state=0)
r.fit(X_train,y_train)
predictions=r.predict(X_test)
r.score(X_test,y_test)
```
